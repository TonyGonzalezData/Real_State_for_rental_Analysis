## ANALYSIS

### Analysis of Rental Price

**What is the average price? And the price range? And by districts? And by neighborhoods?**

**What is the ranking of districts and neighborhoods by average rental price?**

```python
properties_data.m2.describe()
```

```python
properties_data.estimated_rental_price.describe()
```

We can see that there is at least one outlier on the upper end that skews the mean, so we will use the median as a more reliable measure of central tendency.

```python
properties_data.estimated_rental_price.median()
```

```python
properties_data.groupby('district').estimated_rental_price.median().sort_values(ascending = False)
```

We are struck by the maximum estimated rental price.

```python
properties_data.loc[properties_data.estimated_rental_price> 3000]
```

```python
properties_data.loc[properties_data.estimated_rental_price> 3000].m2_disc.unique()
```

```python
properties_data.m2.describe()
```

Rental amounts above 2000 euros correspond to properties larger than 200 m2.

This issue stems from how the rental price was calculated. We used the average price per m2, but this data is not representative at the extremes (properties larger than 140 m2 fall into the highest percentile). For properties with a large number of square meters, the price per m2 typically decreases compared to a smaller property in the same area. Therefore, we need to see how the data evolves based on the m2.

```python
properties_data.groupby('m2_disc').estimated_rental_price.describe()
```

Looking at the existing standard deviations, we can see that the data for properties larger than 150 m2 becomes unreliable. Therefore, we will only consider properties up to 150 m2 in size.

```python
properties_data=properties_data.loc[properties_data.m2 <=150]
```

```python
properties_data.estimated_rental_price.describe()
```

```python
properties_data.m2.describe()
```

**Insight 1: When purchasing a property in the city of La Coruña, the median size is 93 m2 with an estimated rental price of 1068 €. The range starts at a minimum of 393 € and 28 m2, up to a maximum of 150 m2 (limited by data quality) and 2100 €.**

```python
properties_data.groupby('district').estimated_rental_price.agg(['min','max','median'])
```

**Insight 2: The neighborhood with the highest minimum and maximum rental prices is Ensanche.**

**What factors, besides location, determine the rental price? Do m2 influence it? Does the number of rooms have an impact?**

```python
temp=properties_data.groupby('rooms')[['estimated_rental_price','rooms']].median()
```

```python
plt.figure(figsize = (16,8))
sns.lineplot(data = temp, x = 'rooms', y = 'estimated_rental_price')
plt.show();
```

It seems that there is a linear progression as more rooms are added. However, there is a slowdown at 4 rooms, followed by a large gap until 5 rooms. Interestingly, properties with 6 rooms have a lower rental price than those with 5 rooms.

**Insight 3: There is a linear progression between the number of rooms and the rental price, starting to decrease from 5 rooms onwards.** 

Let's perform the same analysis by district.

```python
temp=properties_data.groupby('district')[['estimated_rental_price','rooms']].median()
```

```python
properties_data.groupby('district')[['estimated_rental_price','rooms']].median()
```

```python
plt.figure(figsize = (16,8))
sns.scatterplot(data = temp, x = 'estimated_rental_price', y = 'rooms')
for each in range(0,temp.shape[0]):
    plt.text(temp.estimated_rental_price.iloc[each], temp.rooms.iloc[each], temp.index[each])
plt.show()
```

```python
properties_data.groupby('district')['m2'].median().sort_values(ascending=False)
```

```python
temp=properties_data.groupby('district')[['m2','rooms']].median()
plt.figure(figsize = (25,8))
sns.scatterplot(data = temp, x = 'm2', y = 'rooms')
#Ponemos las etiquetas
for each in range(0,temp.shape[0]):
    plt.text(temp.m2.iloc[each], temp.rooms.iloc[each], temp.index[each])
plt.show()
```

```python
temp=properties_data.groupby('rooms')[['m2','rooms']].median()
plt.figure(figsize = (25,8))
sns.scatterplot(data = temp, x = 'm2', y = 'rooms')
#Ponemos las etiquetas
for each in range(0,temp.shape[0]):
    plt.text(temp.m2.iloc[each], temp.rooms.iloc[each], temp.index[each])
plt.show()
```

There is a clear correlation between the number of rooms and the median rental price by district, internally. In other words, typically, the more square meters, the more rooms, and therefore a higher price.

However, there is disparity between districts in the relationship between the number of rooms and rental price. This is partly due to districts that offer more square meters for the same number of rooms, which results in a higher price. Additionally, the location must be taken into account.

### Analysis of Profitability

**What is the ranking of pre-tax profit per m2 by district?**

```python
temp=properties_data.groupby('district')[['earnings_before_taxes','m2']].mean()
```

```python
(temp.earnings_before_taxes/temp.m2).round(2).sort_values(ascending=False).head(5)
```

**What is the ranking of gross profitability by district?**

```python
temp=properties_data.groupby('district')[['gross_profitability']].mean().sort_values(by='gross_profitability',ascending=False).head(5)
```

```python
plt.figure(figsize = (14,6))
g=sns.barplot(data = temp, x = 'district', y = 'gross_profitability')
vals = g.get_yticks()
g.set_yticklabels(['{:.0%}'.format(val) for val in vals])
plt.show();
```

As we can see, the purchase price impacts the final profitability by district. Only two districts that are in the top 5 for profit/m2 are also in the top 5 for gross profitability by district: Juan Flórez-San Pablo and Paseo de los Puentes-Santa Margarita.

**Insight 1: however, focusing exclusively on final profitability, we should focus on the following districts:**

* **Paseo de los Puentes - Santa Margarita**
* **Juan Flórez - San Pablo**
* **Elviña - A Zapateira**

**What is the ranking of the relationship between profitability and property price by district?**

```python
temp=properties_data.groupby('district')[['sale_price','gross_profitability']].mean()
```

```python
plt.figure(figsize = (18,8))
g=sns.scatterplot(data = temp, x = 'sale_price', y = 'gross_profitability')
for each in range(0,temp.shape[0]):
    plt.text(temp.sale_price.iloc[each], temp.gross_profitability.iloc[each], temp.index[each])
g.set_yticklabels(['{:.0%}'.format(val) for val in vals])
plt.show();
```

**Insight 2: The districts of Elviña - A Zapateira and Paseo de los Puentes are the most interesting for investing in rental housing. They present the best gross profitability/price ratio.**

The questions we must ask ourselves now are:

* What combination of square meters optimizes the profitability ratio in the districts of Elviña - A Zapateira and Paseo de los Puentes?
* Is there any exceptional case in the other districts with a lower average sale price than Elviña that matches its profitability ratio?

```python
temp=properties_data.groupby(['district','m2_disc'])['gross_profitability'].mean().reset_index().sort_values(by='gross_profitability',ascending=False).round(2).head(10)

temp.loc[temp.district.isin(['Elviña - a Zapateira','Paseo de los Puentes-Santa Margarita'])]
```

**Insight 3: Specifically, the most profitable district/square meter combination is Paseo de los Puentes between 51 and 150 m2 and Elviña between 50 and 150 m2.**

```python
#elaboramos la tabla distrito/m2 con sus rentabilidades
temp=properties_data.groupby(['district','m2_disc'])['gross_profitability'].mean().reset_index().sort_values(by='gross_profitability',ascending=False).round(2)
```

```python
#determinamos que distritos tienen un precio de venta inferior al precio de venta medio de Elviña
districts=properties_data.groupby('district').sale_price.mean().sort_values()[:4].reset_index().district.to_list()
```

```python
#calculamos la media de precio de venta de Elviña
mean_profitability=properties_data.loc[properties_data.district=='Elviña - a Zapateira'].gross_profitability.mean().round(2)
```

```python
temp.loc[temp.district.isin(districts) & (temp.gross_profitability >= mean_profitability)]
```

**Insight 4: However, there are two exceptions with a sale price lower than the two top districts that match their profitability:**
* **Os Mallos between 101 and 150 m2**
* **Sagrada Familia between 101 and 150 m2**
